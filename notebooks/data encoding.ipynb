```python
import pandas as pd
from pickle import dump
```

```python
train_df = pd.read_csv("../data/train.csv")
train_df.head()
```

```python
categorical_columns = train_df.select_dtypes('object').columns
numerical_columns = train_df.select_dtypes('int').columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for column in categorical_columns:
    label_encoder = LabelEncoder()
    label_encoder.fit(train_df[column])
    train_df[column] = label_encoder.transform(train_df[column])
    file = open(f'../artifacts/label encoders/{column}.pkl','wb')
    dump(label_encoder,file)
```

```python
train_df.to_csv('../artifacts/label encoded data/label encoded.csv',index=False)
```
